### 각각의 transform 사항 설정하기

```python
train_transform = v2.Compose([ #훈련 데이터는 다양한 변화 주어 요란발짝해보기
    v2.Resize((256, 256)),
    v2.RandomResizedCrop(size=(224, 224), antialias=True),
    v2.RandomHorizontalFlip(p=0.5),
    v2.RandomVerticalFlip(p=0.5),
    v2.ToTensor(),
    v2.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])])

test_transform = v2.Compose([ #테스트 데이터는 데이터 증강 아닌 일관성 평가
    v2.Resize((256, 256)),
    v2.CenterCrop(size=(224, 224)),
    v2.ToTensor(),
    v2.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])])
```

### train 모델 함수로 설정하여 적용

```python
def train(model, data_loader, loss_fn, optimizer, EPOCHS=1): #EPOCHS는 반복의 횟수, 기본값 1 
    loss_history = [] #각 에포크의 손실을 저장할 리스트 
    acc_history = [] #각 에포크의 정확도를 저장할 리스트
    epoch_loss = 0 #step_loss를 참조하여, 한 에포크 동안의 평균 손실. 
    num_corrects = 0
    device = 'cuda' if torch.cuda.is_available() else 'cpu'
    model = model.to(device) #모델을 device로 이동시킴
    for n in range(EPOCHS): #EPOCHS만큼 반복할건데
        step_loss = 0 #현재 에포크에서의 배치별 손실 - 각 배치의 손실을 누적
        num_corrects = 0
        for idx, (X_train, y_label) in enumerate(data_loader): #배치별로 각각 돌릴거임
            inputs = X_train.to(device)   #얘네도 device 저걸로 쓸거야
            labels = y_label.to(device)
            
            optimizer.zero_grad() #배치마다의 기울기 초기화
            outputs = model(inputs)
            loss = loss_fn(outputs, labels)
            loss.backward()
            optimizer.step()
            
            step_loss += loss.item() #현재 배치의 손실 값을 누적함 
            loss_history.append(step_loss) #누적값을 갖다넣음 
            
            train_acc = torch.sum(torch.argmax(outputs, dim=1) == labels) #현재 배치의 예측이 맞은 개수를 계산함
            num_corrects += train_acc #맞춘 갯수를 누적해서 적음
        epoch_avg_loss = sum(loss_history) / len(data_loader)
        epoch_loss=step_loss / len(data_loader) #에포크의 평균 손실
        accuracy = num_corrects / len(data_loader.dataset) #에포크 동안 맞춘 비율
        print('EPOCH {}, loss: {}, acc: {}'.format(n + 1, epoch_loss, accuracy))
```

```python
loss_fn = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters())
```

## requires_grad 속성 

> 훈련 시 : 기울기 계산이 업데이트 되어야 하므로 requires_grad: True

> 조정 시 : 잠궈놓고 해야 하므로 requires_grad: False 하고 조정

> 평가 시 : 기울기 계산 필요 없고, requires_grad 건드리지 말고 평가 함수 따로 작성하여 for 문 앞에 with torch.no_grad(): 사용
